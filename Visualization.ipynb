# Visualization

## Matplotlib
 - https://matplotlib.org/3.1.1/gallery/index.html
 
 

```python
import matplotlib.pyplot as plt
%matplotlib inline 
# set to inline mode (there's also interactive mode)

import numpy as np
```

#### Line Plot

```python
X = np.arange(10)
Y = 2*X*X*X + 3*X*X + 5*X

plt.plot(X,Y, 'g--')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Line Plot')
```

#### Scatter Plot

```python
X = np.arange(10)
Y = np.random.randint(0, 100, len(X))

plt.scatter(X,Y,marker='x', c='r')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Scatter Plot')
```

#### Pie Plot

```python
X = list('ABCDE')
Y = (10,35,20,15,20)

plt.pie(Y,labels=X)
```

#### Bar Plot

```python
X = list('ABCDE')
Y = (10,35,20,15,20)

plt.bar(x=X, height=Y, width=0.5)

plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Bar Plot')
```

#### Stacked Bar Plot

```python
X = list('ABCDE')
Y1 = (10,35,20,15,20)
Y2 = (20,10,10,30,30)

plt.bar(x=X, height=Y1, width=0.5, bottom=None)
plt.bar(x=X, height=Y2, width=0.5, bottom=Y1)

plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Stacked Bar Plot')
```

### Subplots

```python
# 2by2 grid, 1st plot
plt.subplot(2,2,1)

X1 = list('ABCDE')
Y1 = (10,35,20,15,20)

plt.bar(x=X1, height=Y1, width=0.5)


# 2nd plot
plt.subplot(2,2,2)
X2 = list('FGHIJ')
Y2 = (40,35,60,15,20)

plt.bar(x=X2, height=Y2, width=0.5, color='r')

# 3rd plot

plt.subplot(2,2,3)
X3 = list('KLMNO')
Y3 = (20,35,5,35,20)

plt.bar(x=X3, height=Y3, width=0.5, color='g')

# 4th plot
plt.subplot(2,2,4)
X4 = list('PQRST')
Y4 = (5,70,50,10,20)

plt.bar(x=X4, height=Y4, width=0.5, color='y')
```

## Pandas

Pandas built-in visualization tools.

```python
import pandas as pd
```

```python
df = pd.DataFrame(data = np.random.rand(100,2), columns=list('AB'))
df['C'] = 2*df['A']*df['A'] + 3*df['A'] + 5
df['D'] = df['A'] + df['B']
df['E'] = df['A'] > df['B']
df.head()
```

#### Line Plot

```python
df.plot(x='A', y='C')
# lines going back and forward
```

```python
# sorted
df.sort_values(by=['A']).plot(x='A', y='C')
```

#### Scatter Plot

```python
df.plot(x='A',y='C', kind='scatter')
```

#### Bar Plot

```python
# a > b
df.groupby(by='E')['A'].count()
```

```python
# a > b
df.groupby(by='E')['A'].count().plot(kind='bar')
```

```python
# a > b
df.groupby(by='E')['A'].count().plot(kind='pie')
```

#### Histogram

```python
df['A'].plot.hist(bins=10)
```

#### Box Plot

```python
df.plot.box()
```

```python
df['A'].plot.box()
```

#### Area Plot

```python
df.plot.area()
```

## SNS Categorical
**Figure Level :**
- `sns.catplot()`


**Axes Level :**
  - for scatter plot within categories : `stripplot()`, `swarmplot()`
    - for small datasets (too crowded): `kind = strip/swarm` (add small amount of random jitter)
  - for distributions within categories : `boxplot()`, `violinplot()`, `boxenplot()`
    - for larger datasets
  - for statistical estimations within categories : `pointplot()`, `barplot()`, `countplot()`
    - bar : default to **mean** value
    - count : histogram
    - point : plot the point estimate and confidence interval 

```python
sns.catplot(data=tips, x='day', y='total_bill' , kind='strip')
```

```python
sns.catplot(data=tips, x='day', y='total_bill' , kind='swarm')
```

```python
sns.catplot(data=tips, x='day', y='total_bill' , kind='swarm', hue='smoker')
```

```python
sns.catplot(data=tips, x='day', y='total_bill' , kind='swarm', hue='smoker',
           order=['Sat', 'Sun', 'Mon', 'Tue', 'Wed', 'Thur', 'Fri'])
```

```python
sns.catplot(data=tips, x='day', y='total_bill', kind='box')
```

```python
sns.catplot(data=tips, x='day', y='total_bill', kind='box', hue='sex')
```

```python
sns.catplot(data=tips, x='day', y='total_bill', kind='violin')
```

```python
sns.catplot(data=tips, x='day', y='total_bill', kind='violin', inner=None)
```

### Accessing the Axes

```python
g = sns.catplot(data=tips, x='day', y='total_bill', kind='violin', inner=None)

sns.swarmplot(data=tips, x='day', y='total_bill', color='k', size=3, ax=g.ax)
```

```python
sns.catplot(data=titanic, x='sex', y='survived', hue='class', kind='bar')
```

```python
# people in each class (by gender)
sns.catplot(data=titanic, x='sex', hue='class', kind='count')
```

```python
sns.catplot(data=titanic, x='sex', y='survived', hue='class', kind='point')
```

```python
# showing multiple relationship with facets: make multiple axes and plot subsets of the data on each of them
g = sns.catplot(data=tips, x='day', y='total_bill', kind='swarm', hue='smoker',
            row='sex', col='time', height=1.5, aspect=1.5)

g.fig.set_size_inches(15,15)
```
